```python
# Import the file 
crime = pd.read_csv('crimerates-by-state-2005.csv')

#Check the header
crime.head()
```

```python
# Set the figure size
plt.rcParams['figure.figsize'] = [20, 10]
```

### Scatterplot

```python
# Population under 10,000,000
rob = crime[crime['population'] < 10000000]
```

### Bubble Chart

```python
# Remove single outlier of 10 to better see the overall spread
murder =  rob[rob['murder'] < 10]

# Make the scatter plot with regression line
sns.regplot(x = murder.population, y = murder.murder,
            color = 'darkcyan')
plt.title('Murder Rate Per Population Under 10,000,000', 
          size = 30)
plt.xlabel('Population',
          size = 20)
plt.ylabel('Murder rate', 
           size = 20)
plt.show()
```

```python
# Same plot as abobe, with with sozes added for frequency
plt.scatter(murder.population, murder.murder,  
            s =  murder.murder * 1000,
            alpha = 0.6,
            color = 'darkcyan',
            edgecolor = 'darkslategray')
plt.title('Murder Rate Per Population Under 10,000,000', 
          size = 30)
plt.xlabel('Population',
          size = 20)
plt.ylabel('Murder rate', 
           size = 20)

# show the graph
plt.show()
```

### Density Plot

```python
# Make the density plot showing each tye of crime per population
# Uses the fill crime dataset
crime.plot(kind = 'density', 
            subplots = True, 
            layout = (3,3,), 
            sharex = False, 
            title = 'Crime Rates Per Population')

plt.show()
```
