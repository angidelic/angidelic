### Libraries and data file

```python
# Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import the file 
o = pd.read_excel('obama-approval-ratings.xls')

#Check the header
o.head()
```

### Bar chart

```python
# Set graph size
plt.rcParams['figure.figsize'] = [20, 10]

# Assign colors
colors = ['blue', 'firebrick', 'orange']
names = ['Approve', 'Disapprove', 'None']

o.plot(x = 'Issue',
       kind = 'bar',
       stacked = False,
       width = .8,
       color = colors,
       fontsize = 15)

# Plot formatting
plt.legend(fontsize = 15)
plt.xlabel('Issue', 
          fontsize = 20,
          color = 'black')
plt.title('Obama Approval Ratings by Issue', 
          fontsize = 20,
          color = 'black')
```

### Stacked bar chart

```python
# Set graph size
plt.rcParams['figure.figsize'] = [20, 10]

# Assign colors
colors = ['blue', 'firebrick', 'orange']
names = ['Approve', 'Disapprove', 'None']


o.plot(x = 'Issue',
       kind = 'bar',
       stacked = True,
       width = .8,
       color = colors,
       fontsize = 15)

# Plot formatting
plt.legend(fontsize = 15)
plt.xlabel('Issue', 
          fontsize = 20,
          color = 'black')
plt.title('Obama Approval Ratings by Issue', 
          fontsize = 20,
          color = 'black')

```

### Pie chart

```python
# Create new variables of the individual columns
o_sum = o.sum()

a_sum = o_sum['Approve']
b_sum = o_sum['Disapprove']
c_sum = o_sum['None']

o_pie = [a_sum, b_sum, c_sum]
```

```python
# Create labels
o_labels = 'Approve','Disapprove','None'

# Colors are slightly muted compared to the barcharts
colors = ['royalblue', 'indianred', 'khaki']

# Create the pie chart
plt.pie(o_pie, 
        labels = o_labels, 
        autopct = '%1.1f%%', 
        shadow = False, 
        textprops = {'fontsize': 15}, 
        colors = colors)

plt.title('Obama Approval/Disapproval Ratings', 
          size = 20, 
          color = 'black')

# Display the plot
plt.show()
```

### Donut chart

```python
# Create labels
o_labels = 'Approve','Disapprove','None'

# Colors are slightly muted compared to the barcharts
colors = ['royalblue', 'indianred', 'khaki']

# Create the pie chart
plt.pie(o_pie, 
        labels = o_labels, 
        autopct = '%1.1f%%', 
        pctdistance = 0.75,
        shadow = False, 
        textprops = {'fontsize': 18}, 
        colors = colors)

plt.title('Obama Approval/Disapproval Ratings', 
          size = 20, 
          color = 'black')

# Create the hollow middle to draw the donut
circle = plt.Circle((0, 0), 0.60, 
                    fc = 'white')
fig = plt.gcf()
  
# Adding Circle in Pie chart
fig.gca().add_artist(circle)
  

# Displaying Chart
plt.show()
```
