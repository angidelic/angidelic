```python
import math

# Create function for whole numbers
def WholeNum(NumStack):
    FinalNum = 0
    for i in range(0, len(NumStack)):
        FinalNum += (NumStack.pop() * (math.pow(10, i)))
    return FinalNum
```

Design your own CSV parser

In this activity we will be tasked with building our own CSV reader and parser. We will assume that there is no escape character, meaning if you use a comma `,` at any place in your row it means you are starting a new column. We will also assume that the only  function we are interested about is to be able to read a CSV line by line where each read will generate a new dict with the column names as keys and the values for the row as value

A CSV looking like above, when read, will return a dict like this - `{"Name": "Bob", "Age": "24", "Location": "California"}`

### Tasks 

- Open the accompanying file "sales_record.csv" using "r" mode inside a with block and check first that it is opened.
- Read the first line and use string methods to generate a list of all the column names
- start reading the file one by one
- When you read each line pass that line to a function along with the list of the headers
- The work of the function is to construct a dict out of these two and fill up the key:values. Keep in mind that a missing value should result in a None.

```python
from itertools import zip_longest

# Create the function to create the dictionary from the CSV
def CSV_Function(header, line):
    ZippedLine = zip_longest(header, line, fillvalue = None)
    Dict = {kv[0]: kv[1] for kv in ZippedLine}
    return Dict
```

```python
# Read the CSV with the function created above
with open("C:/Users/Madam President/Downloads/Data-Wrangling-with-Python-master/Data-Wrangling-with-Python-master/Lesson02/Activity04/sales_record.csv", "r") as fd:
    FirstLine = fd.readline()
    header = FirstLine.replace("\n", "").split(",")
    for i, line in enumerate(fd):
        line = line.replace("\n", "").split(",")
        d = CSV_Function(header, line)
        print(d)
        if i > 10:
            break
```
